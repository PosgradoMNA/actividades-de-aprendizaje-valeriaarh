##**Análisis de Varianza ANOVA**

Comparación estadística de grupos.
Para analizar variables categóricas.

Este análisis se usa para encontrar la correlación entre los diferentes grupos de una variable categórica.

Se obtienen 2 resultados:
- F-test score: es la variación entre el grupo muestra y se divide por la variación del mismo grupo.
- p-value: grado de confianza.

```python
#con value.counts() se puede saber cuántos elementos hay para
#cada categoría por variable

df['drive-wheels'].value_counts()
```

```python
#Y ponerlo en formato de dataframe
drive_wheels_counts = df['drive-wheels'].value_counts().to_frame()
drive_wheels_counts.rename(columns={'drive-wheels': 'value_counts'}, inplace=True)
drive_wheels_counts.index.name = 'drive-wheels'
drive_wheels_counts
```

```python
#para ver las categorías en una variable se puede
#usar la función .unique()

df['drive-wheels'].unique()
```

```python
df_group_one = df[['drive-wheels','body-style','price']]
#Ver resultado de la agrupación con otras en este caso
#precio y drive-wheels

df_group_one = df_group_one.groupby(['drive-wheels'],as_index=False).mean()
df_group_one
```

```python
# resultado de la agrupación de 3 variables
df_gptest = df[['drive-wheels','body-style','price']]
grouped_test1 = df_gptest.groupby(['drive-wheels','body-style'],as_index=False).mean()
grouped_test1
```

```python
#Como se vio en el video esta tabla no dice mucho,
#asi que se hace el pivoteo de la misma

grouped_pivot = grouped_test1.pivot(index='drive-wheels',columns='body-style')
grouped_pivot
```

```python
#Y se llenan los espacios vacíos o NaN

grouped_pivot = grouped_pivot.fillna(0) #fill missing values with 0
grouped_pivot
```

```python
#PREGUNTA 4: Usar la función groupby para
#las variables 'price' y 'body-style'

df_group_Q4 = df[['price', 'body-style']]
Q4_group = df_group_Q4.groupby(['body-style'],as_index= False).mean()
Q4_group
```

```python
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
fig, ax = plt.subplots()
im = ax.pcolor(grouped_pivot, cmap='RdBu')

#label names
row_labels = grouped_pivot.columns.levels[1]
col_labels = grouped_pivot.index

#move ticks and labels to the center
ax.set_xticks(np.arange(grouped_pivot.shape[1]) + 0.5, minor=False)
ax.set_yticks(np.arange(grouped_pivot.shape[0]) + 0.5, minor=False)

#insert labels
ax.set_xticklabels(row_labels, minor=False)
ax.set_yticklabels(col_labels, minor=False)

#rotate label if too long
plt.xticks(rotation=90)

fig.colorbar(im)
plt.show()
```

```python
from scipy import stats
```

```python
#Coeficiente de Correlación de Pearson entre wheel-base y precio
pearson_coef, p_value = stats.pearsonr(df['wheel-base'], df['price'])
print("The Pearson Correlation Coefficient is", pearson_coef, " with a P-value of P =", p_value)  
```

```python
#Coeficiente de Correlación de Pearson entre horsepower y precio
pearson_coef, p_value = stats.pearsonr(df['horsepower'], df['price'])
print("The Pearson Correlation Coefficient is", pearson_coef, " with a P-value of P = ", p_value)  
```

```python
# ANOVA después de la agrupación de drive-wheels y price
grouped_test2=df_gptest[['drive-wheels', 'price']].groupby(['drive-wheels'])
grouped_test2.head(2)
```

```python
grouped_test2.get_group('4wd')['price']

# ANOVA para 4wd (de la variable drive-wheels) y price
f_val, p_val = stats.f_oneway(grouped_test2.get_group('fwd')['price'], grouped_test2.get_group('rwd')['price'], grouped_test2.get_group('4wd')['price'])  
 
print( "ANOVA results: F=", f_val, ", P =", p_val)   
```
