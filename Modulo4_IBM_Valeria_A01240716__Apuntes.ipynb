Entra el par치metro 'a' o Appending, donde se podr치 escribir l칤neas adicionales sin cambiar los datos iniciales. Ejercicio a continuaci칩n

```python
#Se agregan las lineas de C, D y E al archivo inicial Example2, pero sin borrar las l칤neas A y B que ya estaban

# Write a new line to text file

with open('Example2.txt', 'a') as testwritefile:
    testwritefile.write("This is line C\n")
    testwritefile.write("This is line D\n")
    testwritefile.write("This is line E\n")

#y se verifica que se hayan escrito sin quitar la informaci칩n anterior

with open('Example2.txt', 'r') as testwritefile:
    print(testwritefile.read())
```

Existen par치metros adicionales que pueden hacer el Apendant y la escritura de archivos m치s eficiente:
+r: puedes leer y escribir en el fichero/archivo - no lo corrompe

+w: se puede escribir y leer el fichero/archivo - este hay que usarlo con cuidado porque s칤 puede truncar el archivo, ya que sobreescribe los datos que ya exist칤an y los borra

+a: puedes agregar y leer en un archvo/fichero, es capaz de crear un nuevo archivo incluso si no existe.

```python
with open('Example2.txt', 'a+') as testwritefile:
    print("Initial Location: {}".format(testwritefile.tell()))
    
    data = testwritefile.read()
    if (not data):  #si no hay datos regresa un error, pero como si hay datos los muestra
            print('Read nothing') 
    else: 
            print(testwritefile.read())
            
    testwritefile.seek(0,0) # move 0 bytes from beginning.
    
    print("\nNew Location : {}".format(testwritefile.tell()))
    data = testwritefile.read()
    if (not data): 
            print('Read nothing') 
    else: 
            print(data)
    
    print("Location after read: {}".format(testwritefile.tell()) )
```

```python
#Aqu칤 se hace un ejemplo del uso de +r donde puedes leer y escribir al mismo tiempo
with open('Example2.txt', 'r+') as testwritefile: 
    data = testwritefile.readlines()
    testwritefile.seek(0,0) #write at beginning of file
   
    testwritefile.write("Line 1" + "\n")
    testwritefile.write("Line 2" + "\n")
    testwritefile.write("Line 3" + "\n")
    testwritefile.write("finished\n")
    
    testwritefile.truncate()
    testwritefile.seek(0,0)
    print(testwritefile.read())
```

Tambi칠n se puede hacer copiar la informaci칩n de un archivo a otro al leerlo primero y despu칠s escribir lo que ten칤as en ese en el otro as칤 juntando la informaci칩n de antes.

OJO: se agrega la informaci칩n del otro archivo no la sobreescribe.

```python
#Se toma la informaci칩n trabajada en el Example2.txt y se agrega/escribe en el Example3.txt dejando su informaci칩n original
with open('Example2.txt','r') as readfile:
    with open('Example3.txt','w') as writefile:
          for line in readfile:
                writefile.write(line)

with open('Example3.txt','r') as testwritefile:
    print(testwritefile.read())
```

Con los datos actuales currentMem, se necesita remover los miembro que tienen un 'no' como activos pero llevando el registro de quienes son los miembros que han sido removidos y agregarlos en el exMem

```python
#C칩digo ya dado para correr 

from random import randint as rnd

memReg = 'members.txt'
exReg = 'inactive.txt'
fee =('yes','no')

def genFiles(current,old):
    with open(current,'w+') as writefile: 
        writefile.write('Membership No  Date Joined  Active  \n')
        data = "{:^13}  {:<11}  {:<6}\n"

        for rowno in range(20):
            date = str(rnd(2015,2020))+ '-' + str(rnd(1,12))+'-'+str(rnd(1,25))
            writefile.write(data.format(rnd(10000,99999),date,fee[rnd(0,1)]))


    with open(old,'w+') as writefile: 
        writefile.write('Membership No  Date Joined  Active  \n')
        data = "{:^13}  {:<11}  {:<6}\n"
        for rowno in range(3):
            date = str(rnd(2015,2020))+ '-' + str(rnd(1,12))+'-'+str(rnd(1,25))
            writefile.write(data.format(rnd(10000,99999),date,fee[1]))


genFiles(memReg,exReg)
```

Aqu칤 se hace la limpieza de los datos tal como lo pide la instrucci칩n

```python
def cleanFiles(currentMem,exMem): #se define la funci칩n para hacer la limpieza que involucra las variables dadas
    with open(currentMem,'r+') as writeFile: #Se lee y se escribe en el mismo archivo sin truncarlo
        with open(exMem,'a+') as appendFile: #Se crea un nuevo archivo y se lee
            
            writeFile.seek(0)
            members = writeFile.readlines()
            
            header = members[0] #Se remueven los headers del archivo ya que no son relevantes para los datos
            members.pop(0)
                
            inactive = [member for member in members if ('no' in member)] #Se define cu치les son los miebros activos y cuando 'no'
            
            writeFile.seek(0) 
            writeFile.write(header)

            #se hace el loop para saber cu치les miembros est치n o no activos y se escribe en el archivo correspondiente haciendo la filtraci칩n de cada uno
            for member in members: 
                if (member in inactive):
                    appendFile.write(member)
                else:
                    writeFile.write(member)      
            writeFile.truncate()
                
memReg = 'members.txt'
exReg = 'inactive.txt'
cleanFiles(memReg,exReg)

# Se despliega la informaci칩n de cada archivo de los miembros activos o no

headers = "Membership No  Date Joined  Active  \n"

with open(memReg,'r') as readFile:
    print("Active Members: \n\n")
    print(readFile.read())
    
with open(exReg,'r') as readFile:
    print("Inactive Members: \n\n")
    print(readFile.read())
```
