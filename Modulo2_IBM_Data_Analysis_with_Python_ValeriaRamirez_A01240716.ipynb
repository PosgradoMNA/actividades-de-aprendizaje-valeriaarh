```python
import pandas as pd
import matplotlib.pylab as plt
```

```python
filename = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DA0101EN-SkillsNetwork/labs/Data%20files/auto.csv"
```

```python
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
```

```python
df = pd.read_csv(filename, names = headers)
df.head() # Despliega el data set
```

```python
#Identicicación y reemplazo de valores faltantes
import numpy as np

# Reemplaza "?" a --> NaN
df.replace("?", np.nan, inplace = True)
df.head(5)
```

```python
#Cómo detectar si hay datos perdidos
missing_data = df.isnull()
missing_data.head(5)
```

```python
#Conteo de datos perdidos
for column in missing_data.columns.values.tolist():
    print(column)
    print (missing_data[column].value_counts())
    print("")
```

```python
#Trabajar con datos perdidos
#Calcular Media para la columna de "normalized-losses"

avg_norm_loss = df["normalized-losses"].astype("float").mean(axis=0)
print("Average of normalized-losses:", avg_norm_loss)
```

```python
#Trabajar con datos perdidos
#Reemplazar valores NaN de la columna de "normalized-losses"
df["normalized-losses"].replace(np.nan, avg_norm_loss, inplace=True)
```

```python
#Trabajar con datos perdidos
#Calcular Media para la columna de "bore"

avg_bore=df['bore'].astype('float').mean(axis=0)
print("Average of bore:", avg_bore)
```

```python
#Trabajar con datos perdidos
#Reemplazar valores NaN de la columna de "bore"

df["bore"].replace(np.nan, avg_bore, inplace=True)
```

```python
#PREGUNTA 1: Reemplazar valores NaN de la columna stroke con valores de media

avg_stroke=df['stroke'].astype('float').mean(axis=0)
print("Average of stroke:", avg_stroke)

df["stroke"].replace(np.nan, avg_stroke, inplace = True)
```

```python
#Trabajar con datos perdidos
#Reemplazar valores NaN de la columna horsepower con valores de media

avg_horsepower = df['horsepower'].astype('float').mean(axis=0)
print("Average horsepower:", avg_horsepower)

df['horsepower'].replace(np.nan, avg_horsepower, inplace=True)
```

```python
#Trabajar con datos perdidos
#Reemplazar valores NaN de la columna peak-rpm con valores de media

avg_peakrpm=df['peak-rpm'].astype('float').mean(axis=0)
print("Average peak rpm:", avg_peakrpm)

df['peak-rpm'].replace(np.nan, avg_peakrpm, inplace=True)
```

```python
#Visualizar los valores presentes en una columna específica
#En este caso número de puertas

df['num-of-doors'].value_counts()
```

```python
#De lo anterior, cuál es el que repite más veces

df['num-of-doors'].value_counts().idxmax()
```

```python
#Reemplazo de valores perdidos en columna num-of-doors por la moda

df["num-of-doors"].replace(np.nan, "four", inplace=True)
```

```python
#Remover renglones que no tienen datos de precio

df.dropna(subset=["price"], axis=0, inplace=True)

# Reset del index por los renglones removidos
df.reset_index(drop=True, inplace=True)
```

```python
df.head() #Despliegue del set de datos sin valores perdidos
```

```python
#Tipo de datos presente en cada columna

df.dtypes
```

```python
#Hay columnas que su tipo de datos desplegado no es el que corresponde
#Se hacer conversión de datos correspondiente

df[["bore", "stroke"]] = df[["bore", "stroke"]].astype("float")
df[["normalized-losses"]] = df[["normalized-losses"]].astype("int")
df[["price"]] = df[["price"]].astype("float")
df[["peak-rpm"]] = df[["peak-rpm"]].astype("float")
```

```python
#Despliegue de tipo de datos ya corregido

df.dtypes
```

```python
df.head()
```

```python
#Estandarización de datos
#Ejemplo la mayoría de los valores están en términos Galones/Milla
#pero hay valores presentes en Litros/Kilómetro
#se aplican fórmulas de conversión


df['city-L/100km'] = 235/df["city-mpg"]
df.head()
```

```python
#PREGUNTA 2: Transforma mpg a L/100km en la columna 
#"highway-mpg" y cambia el nombre de la columna a "highway-L/100km"

df['highway-L/100km'] = 235/df["highway-mpg"]
df.head()
```

```python
#Normalización de datos
#Tranformación de datos al mismo rango de las columnas Lenght y Width

df['length'] = df['length']/df['length'].max()
df['width'] = df['width']/df['width'].max()
```

```python
#PREGUNTA 3: Normalizar la columna height

df['height'] = df['height']/df['height'].max()
```

```python
#Se muestra la tabla de variables length, width y height
#con valores normalizados entre 0 y 1

df[["length","width","height"]].head()
```

```python
#Binning o Agrupación de datos
#Proceso de transformar variables continuas a grupos discretizados
#para análisis

#Ejemplo con la columna horsepower

df["horsepower"]=df["horsepower"].astype(int, copy=True)
```

```python
#Graficación

%matplotlib inline
import matplotlib as plt
from matplotlib import pyplot
plt.pyplot.hist(df["horsepower"])

# set x/y labels and plot title
plt.pyplot.xlabel("horsepower")
plt.pyplot.ylabel("count")
plt.pyplot.title("horsepower bins")
```

```python
#Tiene muchos bins, hacemos otra agrupación

bins = np.linspace(min(df["horsepower"]), max(df["horsepower"]), 4)
bins
```

```python
#se asignan nombres a los grupos:

group_names = ['Low', 'Medium', 'High']
```

```python
#Función cut para determina que valores del dataframe de horsepower
#corresponde a qué grupo

df['horsepower-binned'] = pd.cut(df['horsepower'], bins, labels=group_names, include_lowest=True )
df[['horsepower','horsepower-binned']].head(20)
```

```python
#conteo por grupo

df["horsepower-binned"].value_counts()
```

```python
%matplotlib inline
import matplotlib as plt
from matplotlib import pyplot
pyplot.bar(group_names, df["horsepower-binned"].value_counts())

# set x/y labels and plot title
plt.pyplot.xlabel("horsepower")
plt.pyplot.ylabel("count")
plt.pyplot.title("horsepower bins")
```

```python
%matplotlib inline
import matplotlib as plt
from matplotlib import pyplot


# draw historgram of attribute "horsepower" with bins = 3
plt.pyplot.hist(df["horsepower"], bins = 3)

# set x/y labels and plot title
plt.pyplot.xlabel("horsepower")
plt.pyplot.ylabel("count")
plt.pyplot.title("horsepower bins")
```

```python
#Variable indicador o Dummy
#Variable numérica para etiquetar categorías
#Se muestran los nombres de las columnas del data set

df.columns
```

```python
#Variable 1 dummy para agreupar al diesel y gasolina

dummy_variable_1 = pd.get_dummies(df["fuel-type"])
dummy_variable_1.head()
```

```python
dummy_variable_1.rename(columns={'gas':'fuel-type-gas', 'diesel':'fuel-type-diesel'}, inplace=True)
dummy_variable_1.head()
```

```python
df = pd.concat([df, dummy_variable_1], axis=1)
df.drop("fuel-type", axis = 1, inplace=True)
df.head()
```

```python
#PREGUNTA 4: Crear variable dummy para la columna aspiración
dummy_variable_2 = pd.get_dummies(df['aspiration'])

#Se tienen 2 valores en esta columna: std y turbo
dummy_variable_2.rename(columns={'std':'aspiration-std', 'turbo': 'aspiration-turbo'}, inplace=True)
dummy_variable_2.head()
```

```python
#PREGUNTA 5: Unir la variable dummy al data set original

df = pd.concat([df, dummy_variable_2], axis=1)
df.drop("aspiration", axis = 1, inplace=True)
df.head()
```

```python
#Guardar nuevo data set ya limpio

df.to_csv('clean_df.csv')
```
