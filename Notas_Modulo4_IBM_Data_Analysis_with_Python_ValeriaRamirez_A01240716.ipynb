##**Modelo** -
Puede ser pensado como una ecuación matemática que se usa para predecir un valor, con otros valores ya dados. Y relacionar una o más variables independientes a variables dependientes.

Entre más datos relevantes tienes, más preciso será tu modelo.

##**Regresión Lineal**

**Regresión Lineal Simple**

Una variable independiente para hacer una predicción.
Se tiene un set de training point para entrenar el modelo. Y se puede usar el modelo para predecir puntos que no se han visto.


**Regresión Lineal Múltiple**

Múltiples variables independientes para hacer una predicción.

##**Measures for In-Sample Evaluation**

Esto es una manera de determinar que tan bien se ajusta el modelo al dataset.

Y dos medidas importantes para esto son:

- Mean Squared Error (MSE)
  Diferencias entre y y y_hat, se hace el cuadrado de eso y se divide entre en número de muestras.

  
- R-squares(R^2)

Coeficiente de determinación, e indica que tan cerca están los datos de la línea de regresión ajustada. Es el porcentaje de variación del target (Y) que se explica por el modelo lineal.

En resumen, es como comparar un modelo de regresión a un modelo simple.

```python
#Regresión polinomial

def PlotPolly(model, independent_variable, dependent_variabble, Name):
    x_new = np.linspace(15, 55, 100)
    y_new = model(x_new)

    plt.plot(independent_variable, dependent_variabble, '.', x_new, y_new, '-')
    plt.title('Polynomial Fit with Matplotlib for Price ~ Length')
    ax = plt.gca()
    ax.set_facecolor((0.898, 0.898, 0.898))
    fig = plt.gcf()
    plt.xlabel(Name)
    plt.ylabel('Price of Cars')

    plt.show()
    plt.close()
```

```python
x= df['highway-mpg']
y= df['price']
```

```python
#Se usa la función polinomial al cubo
f = np.polyfit(x, y, 3)
p = np.poly1d(f)
print(p)
```

```python
PlotPolly(p, x, y, 'highway-mpg')
```

```python
np.polyfit(x, y, 3)
```

```python
#PREGUNTA 4: Crea un modelo de nivel 11 con las variables anteriores:

f1 = np.polyfit(x, y, 11)
p1 = np.poly1d(f)
print(p1)

PlotPolly(p1,x,y, 'Highway MPG')
```
