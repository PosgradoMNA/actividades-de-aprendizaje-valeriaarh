**Lab3: HTTP and Requests**


```python
#Importación de librerías
import requests

import os 
from PIL import Image
from IPython.display import IFrame

#Se hace el request con el comando GET

url='https://www.ibm.com/'
r=requests.get(url)

#Se imprime el request
r.status_code
print(r.request.headers)
print("request body:", r.request.body)

#Se imprimen los headers
header=r.headers
print(r.headers)

#Se imprime el header de Fecha 'Date'
header['date']

header['Content-Type']

r.encoding
```

```python
#Se puede cargar diferentes tipos de datos que no sean relacionados a texto, como imágenes:

url='https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0101EN-SkillsNetwork/IDSNlogo.png'
r=requests.get(url)

print(r.headers)

#Se puede ver la información de la imagen
path=os.path.join(os.getcwd(),'image.png')
path

#Se guarda el el archivo para poder accesarlo después, primero se abre con el formato 'wb' = write binary
with open(path,'wb') as f:
    f.write(r.content)

#y se imprime la imagen
Image.open(path)  
```

```python
#Cómo accesar a los parámetros de un URL
url_get='http://httpbin.org/get'

#Se crea un Query tipo string y se agrega a un diccionario. en este caso los Keys
payload={"name":"Joseph","ID":"123"}
r=requests.get(url_get,params=payload)
print(r.url) #Se imprime el URL con los parámetros establecidos

print(r.text)
```

```python
#Cómo postear los Requests

url_post='http://httpbin.org/post'
r_post=requests.post(url_post,data=payload)

print("POST request URL:",r_post.url ) #se imprime el URL con el parámetro establecido
print("GET request URL:",r.url) #Se compara con el parámetro del ejercicio anterior y establece la diferencia entre GET y POST
```
